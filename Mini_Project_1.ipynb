```python
import pandas as pd

print("====================================="*3)
print("STUDENT PERFORMANCE EXPLORER REPORT")
print("====================================="*3)

df=pd.read_csv('student_performance.csv')

print("[1] DATASET OVERVIEW")
print("---"*40)
# print("Total Students: ", len(df))
print("Total Students: ", df.shape[0])
# print("Total Columns: ", len(df.columns))
print("Total Students: ", df.shape[1])

print("Departments: ", len(df['department'].unique()))
```

```python
print("[2] DEPARTMENT-WISE STUDENT COUNT")
print("---"*40)
# print(df.groupby('department').count())  -> lists cout of all columns
print(df.groupby('department')['name'].count().sort_values(ascending=False))
```

```python
print("[3] SCORE ANALYSIS")
print("-"*40)
print("Highest Math Score:", df['math_score'].max())
print("Lowest Math Score:", df['math_score'].min())
print(f"Average Math Score: {df['math_score'].mean():.2f}")
print(f"Average Programming Score: {df["programming_score"].mean():.2f}") #instead of round(2) -> ":.2f " -> bu this must be inside f"{:.2f}"
print(f"Average Attendance: {df['attendance_percentage'].mean():.2f}%")
```

```python
print("[4] GENDER-WISE AVERAGE SCORES")
print("-"*40)
df['total_scores']=df['math_score']+df['programming_score']+df['english_score']+df['science_score']
print(df.groupby('gender')['total_scores'].mean().round(2))
```

```python
print("[5] TOP 5 PERFORMERS")
print("-"*40)
top5 = df.sort_values('total_scores', ascending=False).head(5)
for index, row in top5.iterrows():
  print(f"{row['name']:<20} | {row['department']:<18} | Score: {row['total_scores']}")
```

```python
print("[6] STUDENTS WITH LOW ATTENDANCE (Below 75%)")
print("-"*40)
low_att=df[df['attendance_percentage']<75][['name', 'department', 'attendance_percentage']]
if len(low_att)>0:
  print(low_att.to_string(index=False)) #index=False -> eliminates index from printing

else:
  print("All students have attendance above 75%")
print("\n"+"="*60)
print("Report Complete")
print("="*60)
```
